# 순이동률 순위

```python
# 1. 계산에 사용할 연도 컬럼 리스트 추출
year_cols = [col for col in df_filtered.columns if '순이동률' in col]

# 2. 연도별 순위 기반 점수 매기기
for col in year_cols:
    # 순위 계산 (값이 클수록 1위)
    ranks = df_filtered[col].rank(ascending=False, method='min')
    # .astype(int)로 소수점 제거 (2.0 -> 2)
    df_filtered.loc[:, f'{col}_점수'] = (8 - ranks).astype(int)

# 3. 점수 총합 계산 및 정수 변환
score_cols = [col for col in df_filtered.columns if '_점수' in col]
df_filtered.loc[:, '점수총합'] = df_filtered[score_cols].sum(axis=1).astype(int)

# 4. 결과 확인 (점수총합 기준 내림차순 정렬 추가)
final_cols = ['행정기관'] + score_cols + ['점수총합']
# ascending=False가 높은 순서(내림차순) 정렬입니다.
df_filtered[final_cols].sort_values(by='점수총합', ascending=False).style.hide(axis='index')
```

# 출생률

```python
birth_rate = pd.read_csv("birth_rate.csv", index_col="광역시별")
birth_rate
```

```python
# 광역시 연도별 출생률 변화 코드 

from matplotlib import rc
rc("font", family="malgun gothic")
plt.figure(figsize=(10, 6))

for region in birth_rate.index:
    plt.plot(
        birth_rate.columns,
        birth_rate.loc[region],
        marker='o',
        label=region
    )

plt.title("시도별 연도별 출생률 변화 (2015~2024)")
plt.xlabel("연도")
plt.ylabel("출생률")
plt.xticks(birth_rate.columns)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# 광역시 연도별 출생률 수준 vs 감소율 비교

import matplotlib.pyplot as plt
import pandas as pd

for year in range(2015, 2025):

    result = pd.DataFrame(index=birth_rate.index)

    # 🔹 해당 연도 출생률 (수준)
    result["출생률"] = birth_rate[str(year)]

    # 🔹 최대 → 최저 감소율 (%): 연도와 무관하므로 그대로
    result["감소율(%)"] = (
        (birth_rate.max(axis=1) - birth_rate.min(axis=1))
        / birth_rate.max(axis=1)
    ) * 100

    # 🔹 산점도
    plt.figure(figsize=(9, 7))
    plt.scatter(
        result["출생률"],
        result["감소율(%)"]
    )

    # 시도명 라벨
    for region in result.index:
        plt.text(
            result.loc[region, "출생률"] + 0.005,
            result.loc[region, "감소율(%)"] + 0.3,
            region,
            fontsize=9
        )

    plt.xlabel(f"{year}년 출생률 (수준)")
    plt.ylabel("출생률 감소율 (%)")
    plt.title(f"{year}년 시도별 출생률 수준 vs 감소율 비교")

    # 🔹 연도별 중앙값 기준선
    plt.axvline(result["출생률"].median(), linestyle='--')
    plt.axhline(result["감소율(%)"].median(), linestyle='--')

    plt.grid(True)
    plt.savefig("2015~2024 4분면.png")
    plt.show()
```

# 출생률 점수 종합 계산 

```python
# 4분면 점수 총합 계산

import pandas as pd
import numpy as np

def 사분면_점수_총합_계산(
    birth_rate: pd.DataFrame,
    시작연도=2015,
    종료연도=2024,
    기준연도=2015
):
    # 🔹 연도 컬럼 정수로 통일
    데이터 = birth_rate.copy()
    데이터.columns = 데이터.columns.astype(int)

    시작연도 = int(시작연도)
    종료연도 = int(종료연도)
    기준연도 = int(기준연도)

    연도목록 = list(range(시작연도, 종료연도 + 1))

    # 🔹 기준연도 출생률
    기준출생률 = 데이터[기준연도].replace(0, np.nan)

    # 🔹 결과 저장용
    결과 = pd.DataFrame(index=데이터.index)

    # 🔹 연도별 4분면 점수 계산
    for 연도 in 연도목록:
        # X축: 해당 연도 출생률(수준)
        출생률_수준 = 데이터[연도]

        # Y축: 기준연도 대비 감소율(%)
        출생률_감소율 = (기준출생률 - 출생률_수준) / 기준출생률 * 100

        # 중앙값 기준선
        수준_중앙값 = 출생률_수준.median()
        감소율_중앙값 = 출생률_감소율.median()

        # 🔸 경계값 = 위험 (보수적 규칙)
        낮은_출생률 = 출생률_수준 <= 수준_중앙값
        빠른_감소 = 출생률_감소율 >= 감소율_중앙값

        # 🔹 4분면 점수 부여
        점수 = []
        for 지역 in 데이터.index:
            if 낮은_출생률.loc[지역] and 빠른_감소.loc[지역]:
                점수.append(4)   # 좌상단
            elif (not 낮은_출생률.loc[지역]) and 빠른_감소.loc[지역]:
                점수.append(3)   # 우상단
            elif 낮은_출생률.loc[지역] and (not 빠른_감소.loc[지역]):
                점수.append(2)   # 좌하단
            else:
                점수.append(1)   # 우하단

        결과[f"{연도}년_점수"] = 점수

    # 🔹 총합 및 보조 지표
    결과["누적_위험점수"] = 결과.sum(axis=1)
    결과["평균_출생률(2015~2024)"] = 데이터[연도목록].mean(axis=1)
    결과["출생률_감소율_2015→2024(%)"] = (
        (기준출생률 - 데이터[종료연도]) / 기준출생률 * 100
    )

    # 🔹 위험도 기준 정렬
    결과 = 결과.sort_values(
        ["누적_위험점수", "출생률_감소율_2015→2024(%)"],
        ascending=False
    )

    return 결과

사분면_위험점수표 = 사분면_점수_총합_계산(
    birth_rate=birth_rate,
    시작연도=2015,
    종료연도=2024,
    기준연도=2015
)

사분면_위험점수표
```
